```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer, ENGLISH_STOP_WORDS
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd
import re
```

```python
df = pd.read_csv("spam_email.csv", sep='\t', names=['label','message'])
df.head(8)
```

```python
#Dataset preprocessing

df['label'] = df['label'].map({'ham': 0, 'spam': 1})
df.head()
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z0-9 ]', '', text)
    words = text.split()
    words = [w for w in words if w not in ENGLISH_STOP_WORDS]
    return " ".join(words)
df['message'] = df['message'].apply(clean_text)
df.head()

X = df['message']
y = df['label']

tfidf = TfidfVectorizer(max_features=3000, ngram_range=(1,2))
X_tfidf = tfidf.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y, test_size=0.2, random_state=42
)
lr = LogisticRegression(max_iter=1000)
svm = SVC(kernel='linear')

lr.fit(X_train, y_train)
svm.fit(X_train, y_train)

lr_pred = lr.predict(X_test)
svm_pred = svm.predict(X_test)

print("Logistic Regression Accuracy:", accuracy_score(y_test, lr_pred))
print("SVM Accuracy:", accuracy_score(y_test, svm_pred))
```

```python
print("\n--- Logistic Regression Report ---")
print(classification_report(y_test, lr_pred))

print("\n--- SVM Report ---")
print(classification_report(y_test, svm_pred))
```

```python
if accuracy_score(y_test, lr_pred) > accuracy_score(y_test, svm_pred):
    final_model = lr
    print("Final Model: Logistic Regression")
else:
    final_model = svm
    print("Final Model: SVM")
    
def predict_spam(text):
    text = clean_text(text)
    vector = tfidf.transform([text])
    result = final_model.predict(vector)
    return "Spam" if result[0] == 1 else "Not Spam"
print(predict_spam("Congratulations! You won a free ticket"))
print(predict_spam("Hey let's meet tomorrow"))
print(predict_spam("Come meet tomorrow"))
```
